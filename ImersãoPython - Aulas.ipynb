```python
import pandas as pd
import plotly.express as px
```

```python
#coleta de dados e impressão dos mesmos
df_principal = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name = "Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name = "Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name = "Ticker")
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/Tabela de ações.xlsx", sheet_name = "Chat GPT")
df_chatgpt
```

```python
#selecionando apenas as colunas desejadas
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
#renomeando as colunas
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_Final', 'Var. Dia (%)':'var_dias_pct'}).copy()
df_principal
```

```python
#criando uma nova coluna com o valor em percentual e o valor inicial da ação no dia
df_principal['Var_pct'] = df_principal['var_dias_pct'] / 100
df_principal['valor_inicial'] = df_principal['Valor_Final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
#fazendo o merge das duas planilhas diferentes, baseado nos dados iguais
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
#retirando a coluna igual
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
#criando a coluna com o valor de variação em R$
df_principal['Variacao_rs'] = (df_principal['Valor_Final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
#formatando as saídas
pd.options.display.float_format="{:.2f}".format
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
#renomeando as colunas
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_teorica'}).copy()
df_principal
```

```python
#criando a coluna de resultados usando a lógica do procv
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estavel'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Idade (em anos)'].apply(lambda x: 'Mais de 100' if x > 0 else ('Menos de 50' if x < 0 else 'Entre 50 e 100'))
df_principal
```

```python
#COMEÇO DAS ANÁLISES DESSA TABELA

#Maior valor
maior = df_principal['Variacao_rs'].max()

#Menor Valor
menor = df_principal['Variacao_rs'].min()

#Media
media = df_principal['Variacao_rs'].mean()

#Média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

#Média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

#Impressão dos resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais x Resultado')
fig.show()
```
